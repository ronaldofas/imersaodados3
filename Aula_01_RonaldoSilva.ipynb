# Desafio 01: Investigar por que a classe tratamento é tão desbalanceada?

O desbalanceamento da coluna tratamento se deve ao tipo de pesquisa desenvolvida que gerou a base de dados em estudo. A pesquisa foi desenvolvida utilizando a metodologia Grupo de Controle, onde uma parte do grupo de análise é tratado com um placebo ou não é tratada, sendo simplesmente acompnhado para forneceber base de comparação para o restante da amostra.

Fonte: https://pt.wikipedia.org/wiki/Grupo_de_controle#:~:text=Um%20grupo%20de%20controle%20cient%C3%ADfico,dois%20experimentos%20id%C3%AAnticos%20s%C3%A3o%20conduzidos.

#Desafio 03: Proporção das classes tratamento.




````python
tratamento_com_drogas = dados['tratamento'].value_counts()[0]
tratamento_sem_drogas = dados['tratamento'].value_counts()[1]
dados_base = dados.shape
qtd_total = dados_base[0]
proporcao_com_drogas = tratamento_com_drogas / qtd_total * 100
proporcao_sem_drogas = tratamento_sem_drogas / qtd_total * 100
print('A proporção de tratamentos com droga é de {0}% e sem drogas é de {1}%.'.format(round(proporcao_com_drogas, 2), round(proporcao_sem_drogas, 2)))
````

#Desafio 06: Renomear as colunas tirando o hífen.



````python
for coluna in dados.columns:
  if '-' in coluna:
    dados.rename(columns={coluna : coluna.replace('-', '')}, inplace = True)
````

#Desafio 07: Deixar os gráficos bonitões (Título, label, etc.) Matplotlib.pyplot).
