# Desafio 03: Plotar os histogramas com seaborn.

```python
# dados['g-0'].hist(bins=100)
fig = plt.figure(figsize=(10,8))
fig.suptitle('Distribuição do gene g-0', fontsize=20)
plt.xlabel('Resultados', fontsize=18)
plt.ylabel('Quantidade g-0', fontsize=18)
ax = sns.histplot(x='g-0', data=dados, bins=100)
plt.show()
```

```python
# dados['g-19'].hist(bins=100)
fig = plt.figure(figsize=(10,8))
fig.suptitle('Distribuição do gene g-19', fontsize=20)
plt.xlabel('Resultados', fontsize=18)
plt.ylabel('Quantidade g-19', fontsize=18)
ax = sns.histplot(x='g-19', data=dados, bins=100)
plt.show()
```

```python
# dados.loc[:, 'g-0':'g-771'].describe().T['mean'].hist(bins=30)
fig = plt.figure(figsize=(10,8))
fig.suptitle('Distribuição média dos genes', fontsize=20)
plt.xlabel('Resultados', fontsize=18)
plt.ylabel('Quantidades', fontsize=18)
ax = sns.histplot(data=dados.loc[:, 'g-0':'g-771'].describe().T['mean'], bins=50)
plt.show()
```

```python
# dados.loc[:, 'g-0':'g-771'].describe().T['min'].hist(bins=30)
fig = plt.figure(figsize=(10,8))
fig.suptitle('Distribuição de valor mínimo dos genes', fontsize=20)
plt.xlabel('Resultados', fontsize=18)
plt.ylabel('Quantidades', fontsize=18)
ax = sns.histplot(data=dados.loc[:, 'g-0':'g-771'].describe().T['min'], bins=20)
plt.show()
```

```python
# dados.loc[:, 'g-0':'g-771'].describe().T['max'].hist(bins=30)
fig = plt.figure(figsize=(10,8))
fig.suptitle('Distribuição de valor máximo dos genes', fontsize=20)
plt.xlabel('Resultados', fontsize=18)
plt.ylabel('Quantidades', fontsize=18)
ax = sns.histplot(data=dados.loc[:, 'g-0':'g-771'].describe().T['max'], bins=20)
plt.show()
```

# Estudar as estatísticas retornadas pelo método .describe().

Count = contagem do total de dados para cada coluna da tabela.
mean = Média daquela coluna calculada através da soma do valores por coluna dividido pelo contagem dos dados da coluna (count)
std = Medida utilizada para demonstrar o quanto o conjunto de dados (coluna) é homogênio.
min = valor minimo da coluna
25% = quartil menor que demonstra que desse ponto para baixo existem 25% dos valores da amostra e desse ponto para cima existem 75% dos valores da amostra.
50% = quartil média que demonstra que desse ponto para baixo existem 50% dos valores da amostra e desse ponto para cima existem 50% dos valores da amostra.
75% = quartil superior que demonstra que desse ponto para baixo existem 75% dos valores da amostra e desse ponto para cima existem 25% dos valores da amostra.
max = valor máximo da coluna

```python
plt.figure(figsize=(10,8))
fig.suptitle('Aplicação do tratamento para o c-0', fontsize=20)
plt.xlabel('Tratamento', fontsize=18)
plt.ylabel('Grupo c-0', fontsize=18)
sns.boxplot(x='c-0', y='tratamento', data=dados)
plt.show()
```
